```python
# create a dot file of concern connections for the paper
G = nx.Graph()
G.add_nodes_from(results['PConcern'])

# start counting links between concerns
connections = dict()
for index, row in results.iterrows():
    s_concerns = row['SConcern'].split(',')
    for concern in s_concerns:
        if concern != "None":
            if (row['PConcern'], concern) not in connections:
                connections[(row['PConcern'], concern)] = 1
            else:
                connections[(row['PConcern'], concern)] += 1
                
# fold duplicates into each other
new_connections = connections.copy()
for key1, key2 in connections:
    if (key1, key2) in new_connections and (key2, key1) in new_connections:
        new_connections[(key1, key2)] += new_connections[(key2, key1)]
        del(new_connections[(key2, key1)])

# add to the graph as weighted edges
for key1, key2 in new_connections:
    G.add_edge(key1, key2, label=new_connections[(key1, key2)])

# export as a dot file for rendering with graphviz 
nx.drawing.nx_pydot.write_dot(G, 'graph.dot')
```
